#### Simulate a fair die and a biased 6-sided die. The biased die has probabilities {0.15, 0.15, 0.15, 0.15, 0.15, 0.25}. Create a visualization that compares outcomes of multiple rolls of a fair die and this biased die. You can use a single visualization or multiple visualizations to demonstrate the difference in outcomes for the dice. The user of your notebook should be able to alter the number of simulations as an argument to a function.â€‹

```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
def simulation(times):
    fair=np.random.choice([1, 2, 3, 4, 5, 6], size=times, p=[1/6,1/6,1/6,1/6,1/6,1/6])
    unfair=np.random.choice([1, 2, 3, 4, 5, 6], size=times, p=[0.15, 0.15, 0.15, 0.15, 0.15, 0.25])
    return fair,unfair
# Simulating both fair and biased die, with fair die's each side having 1/6 probability and biased die having diff
```

```python
def visualization(dFiar, dBiased):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.hist(dFiar,color='blue', edgecolor='black')
    plt.title("Fair Die")
    plt.xlabel("Die Face")
    plt.ylabel("Frequency")
    
    plt.subplot(1, 2, 2)
    plt.hist(dBiased, color='red', edgecolor='black')
    plt.title("Biased Die")
    plt.xlabel("Die Face")
    plt.ylabel("Frequency")
    
    plt.show()

#Plotting two graphs for fair and biased die
```

```python
#times=int(input('Enter number of simulations:')) #number of times simulation will run
times=1000000
start=time.time()
dFair, dBiased = simulation(times)#get the lists for fair and biased die
print('time taken=',time.time()-start)
visualization(dFair, dBiased)#pass the list to visualization for fair and biased die
```
