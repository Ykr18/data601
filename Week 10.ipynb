Week 6 Simulate a fair die and a biased 6-sided die. The biased die has probabilities {0.15, 0.15, 0.15, 0.15, 0.15, 0.25}. Create a visualization that compares outcomes of multiple rolls of a fair die and this biased die. You can use a single visualization or multiple visualizations to demonstrate the difference in outcomes for the dice. The user of your notebook should be able to alter the number of simulations as an argument to a function


```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

Instead of using a list with 1/6, or 0.15  six times and five times respectively, the operation [1/6]*6 and [0.15]*5 is significantly faster. The original and new timings can be viewed while calling the functions. 

```python
def simulation(times):
    fair=np.random.choice([1, 2, 3, 4, 5, 6], size=times, p=[1/6,1/6,1/6,1/6,1/6,1/6])
    unfair=np.random.choice([1, 2, 3, 4, 5, 6], size=times, p=[0.15, 0.15, 0.15, 0.15, 0.15, 0.25])
    return fair,unfair
    
def simulationNew(times):
    fair = np.random.choice(6, size=times, p=[1/6]*6) #changes made here
    unfair = np.random.choice(6, size=times, p=[0.15]*5 + [0.25]) #changes made here
    return fair, unfair
# Simulating both fair and biased die, with fair die's each side having 1/6 probability and biased die having diff
```

```python
def visualization(dFiar, dBiased):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.hist(dFiar,color='blue', edgecolor='black')
    plt.title("Fair Die")
    plt.xlabel("Die Face")
    plt.ylabel("Frequency")
    
    plt.subplot(1, 2, 2)
    plt.hist(dBiased, color='red', edgecolor='black')
    plt.title("Biased Die")
    plt.xlabel("Die Face")
    plt.ylabel("Frequency")
    
    plt.show()
#Plotting two graphs for fair and biased die
```

```python
#times=1000000 #tested on 1000000 on both notebooks
times=int(input("Enter number of times"))
start=time.time()
dFair, dBiased = simulation(times)
print('time taken old =', time.time() - start)

start=time.time()
dFair, dBiased = simulationNew(times)
print('time taken new=', time.time() - start)

visualization(dFair, dBiased)
```
