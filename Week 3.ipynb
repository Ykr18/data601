```python
import requests
import pandas as pd
import random as rd
import numpy as np
```

### FUNCTIONS TO GENERATE DATA

```python
#Using randommer api to get names, emails,addresses and phone numbers

def getName(en):
    url="https://randommer.io/api/Name"
    headers={"X-Api-Key": "a146a6e221fc49ffb60cc15f59dc2ed6"}
    params={"nameType":"fullname","quantity":en}
    res=requests.get(url, headers=headers, params=params)
    if res.status_code==200:
        name=res.json()
        return name
    else:
        print(f"Error: {res.status_code}")


def getAddr(en):
    headers={"X-Api-Key": "a146a6e221fc49ffb60cc15f59dc2ed6"}
    url="https://randommer.io/api/Misc/Random-Address"
    params={"Culture":'en',"Number":en}
    res = requests.get(url, headers=headers, params=params)
    if res.status_code==200:
        addr=res.json()
        return addr
    else:
        print(f"Error: {res.status_code}")
    
def getNum(en):
    headers={"X-Api-Key": "a146a6e221fc49ffb60cc15f59dc2ed6"}
    url="https://randommer.io/api/Phone/Generate"
    params={"CountryCode": "US","Quantity": en}
    res=requests.get(url, headers=headers, params=params)
    if res.status_code==200:
        num=res.json()
        return num
    else:
        print(f"Error: {res.status_code}")


#The api didnt really have emails so I used the name to get a suggestion from the api and generate a random number and randomly add a domain to it
def getEmail(name):
    domains=["@gmail.com","@icloud.com","@outlook.com","@protonmail.com","@yahoo.com","@zoho.com","@aol.com","@mail.com"]
    headers={"X-Api-Key": "a146a6e221fc49ffb60cc15f59dc2ed6"}
    url="https://randommer.io/api/Name/Suggestions"
    params={"startingWords":name}
    res=requests.get(url, headers=headers, params=params)
    if res.status_code==200:
        name=res.json()
        idx=rd.sample(range(0,1000),1)
        num=rd.sample(range(0,70),1)
        email=name[num[0]]+str(idx[0])+rd.choice(domains) #concatinating everything together
        return email
    else:
        print(f"Error: {res.status_code}")


#For Year born I'm just picking random numbers
def getYear(en):
    num=rd.choices(range(1930,2024),k=en)
    return num

def getKids(en):
    #generating 200 random numbers of exponential distribution
    #np.clip ensures that the numbers generated stays between 0.3 and 6
    #Cant have 0.3 babies so using round to make the whole array a whole number
    #used 0.3 to generate numbers like 0.3 to 0.8 which after rounding becomes 0 or 1, which will maintain randomnesss
    num=np.clip(np.random.exponential(scale=1, size=en),0.3,6)
    num=np.round(num).astype(int)
    return num

def getRent():
    list=['Rent','Own']
    ch=rd.choice(list)
    return ch

#Same logic with tickets and income
def getTicket(en):
    ticket=np.clip(np.random.exponential(scale=1, size=en),0.3,10)
    ticket=np.round(ticket).astype(int)
    return ticket
    
def getIncome(en):
    ticket=np.clip(np.random.exponential(scale=65000, size=en),51900,107000)#using a scale of 65000 means an average of 65k between the ranges
    ticket=np.floor(ticket).astype(int)
    return ticket
```

### Calling all the Functions to generate data

```python
en=int(input("Enter quantity of data to be generated"))
```

```python
names=getName(en)
addr=getAddr(en)
phno=getNum(en)
year=getYear(en)
kids=getKids(en)
ticket=getTicket(en)
inc=getIncome(en)
email=[]
rent=[]
for i in names:
    email.append(getEmail(i))
    rent.append(getRent())


#Takes about a minute to run
```

```python
fData={
    "Name":names,
    "Address":addr,
    "Phone number":phno,
    "Email":email,
    "Year": year,
    "Number of Kids": kids,
    "Rent or Own": rent,
    "Annual Income":inc,
    "Number of Speeding tickets": ticket
}
df=pd.DataFrame(fData)
```
